### Concatenation in numpy

```python
arr1 = np.array([1,2,3])
arr2 = np.array([1,2,3])
arr3 = np.concatenated((arr1,arr2))
arr3
```

### for two dimension array

```python
arr2d1 = np.array([[1,2,3],
                   [2,3,4],
                                                     ### axis 1 is right side    and axis0 is down side 
                   [5,6,7]])
arr2d2 = np.array([[5,6,7],
                   [6,7,8],
                   [8,9,6]])
np.concatenate((arr2d1, arr2d2), axis=0)

```

```python
arr2d1 = np.array([[1,2,3],
                   [2,3,4],
                                                     ### axis 1 is right side    and axis0 is down side 
                   [5,6,7]])
arr2d2 = np.array([[5,6,7],
                   [6,7,8],
                   [8,9,6]])
np.concatenate((arr2d1, arr2d2), axis=1)
```
